```julia
function E_vacuum(; Lx=3, Ly=3, res=10, ω=2π, dpml=0.5)
    N, M = round(Int, res * (Ly + 2dpml)), round(Int, res * (Lx + 2dpml))
    ε = ones(N*M)
    b = zeros(N,M)
    b[N ÷ 2, M ÷ 2] = 1
    b = vec(b)
    A, = Maxwell2d(Lx, Ly, ε, ω; dpml, resolution=res)
    @show norm(A - transpose(A)) / norm(A)
    return b' * (A \ b)
end
```

```julia
Lx = 5
Ly = 5
res = 20 
ω = π
N = M = 120
ρ_init = zeros(N,M)
b = zeros(N,M)
offset = -round(Int, res * (0.1 + 1.5))
b[N÷2 + offset,M÷2] = 1;
```

```julia
function LDOS_Optimize_with_mask(Lx, Ly, ρ, ω, b, mask; dpml=0.5, resolution=20, Rpml=1e-20, ftol=1e-4, max_eval=500, max_mat=11, mat_loss=0)
    ε = (max_mat + mat_loss * im) * ρ .+ 1
    A, x, y = Maxwell2d(Lx, Ly, ε, ω; dpml, resolution, Rpml)
    ε = vec(ε)
    ρ = vec(ρ)
    b = vec(b)
    D² = A + spdiagm(ω^2 .* ε)
    M, N = length(x), length(y)
    LDOS_vals = Float64[]
    omegas = ComplexF64[]
    iter = 1
    
    function LDOS_obj(ρ, grad)
        ε = (max_mat + mat_loss * im) * ρ .+ 1
        A = D² - spdiagm(ω^2 .* ε)
        LDOS, ∇LDOS = ∇_ε_LDOS(A, ω, b)
        grad .= ∇LDOS
        push!(LDOS_vals, LDOS)

        A_now, _, _ = Maxwell2d(Lx, Ly, reshape(ε, N,M), ω; resolution)
        ω₀_now = sqrt(Arnoldi_eig(A_now, ε, ω, vec(b))[1])
        push!(omegas, ω₀_now)
        @show iter, LDOS
        iter += 1
        return LDOS
    end

    ub = vec(mask)
    opt = Opt(:LD_CCSAQ, M * N)
    opt.lower_bounds = 0
    opt.upper_bounds = ub
    opt.ftol_rel = ftol
    opt.maxeval = max_eval
    opt.max_objective = LDOS_obj

    (LDOS_opt, ρ_opt, ret) = optimize(opt, ρ)
    ε_opt = (max_mat + mat_loss * im) * ρ_opt .+ 1
    A_opt, _, _ = Maxwell2d(Lx, Ly, ε_opt, ω; resolution)
    ω₀_opt = sqrt(Arnoldi_eig(A_opt, vec(ε_opt), ω, vec(b))[1])
    Q_opt = -real(ω₀_opt) / 2imag(ω₀_opt)

    @show numevals = opt.numevals # the number of function evaluations
    @show ω₀_opt
    @show Q_opt
    
    return LDOS_opt, reshape(ε_opt, N,M), LDOS_vals, omegas, x, y
end
```

```julia
function mod_LDOS_Optimize_with_mask(Lx, Ly, ρ, ω, b, x₀, mask; dpml=0.5, resolution=20, Rpml=1e-20, ftol=1e-4, max_eval=500, max_mat=11, mat_loss=0)
    ε = (max_mat + mat_loss * im) * ρ .+ 1
    A, x, y = Maxwell2d(Lx, Ly, ε, ω; dpml, resolution, Rpml)
    ε = vec(ε)
    ρ = vec(ρ)
    b = vec(b)
    D² = A + ω^2 .* spdiagm(ε)
    M, N = length(x), length(y)
    mod_LDOS_vals = Float64[]
    mod_omegas = ComplexF64[]
    iter = 1
    
    function mod_LDOS_obj(ρ, grad)
        ε = (max_mat + mat_loss * im) * ρ .+ 1 
        E = spdiagm(ε)
        E⁻¹ = spdiagm(1 ./ ε)
        A = D² - real(ω)^2 .* E
        C = E⁻¹ * A
        LU = lu(C)
        vals, vecs, _ = eigsolve(z -> LU \ z, x₀, 1, :LM, Arnoldi())
        vals = sqrt.(1 ./ vals .+ ω^2)
        ω₀, u₀ = vals[1], vecs[1]
        A₀ = D² - real(ω₀)^2 .* E
        v = A₀ \ b
        w = conj.(v)
        LDOS = -imag(v' * b)
        
        reals = real.(vals)
        imaginaries = imag.(vals)
        if !isempty(grad) 
            ∂LDOS_∂ε = -imag.(real(ω₀)^2 .* w.^2)
            ∂LDOS_∂ω = -imag(2real(ω₀) .* sum(w.^2 .* conj.(ε)))
            ∂ω_∂ε = -ω₀ .* u₀.^2 ./ 2sum(u₀.^2 .* ε)
            ∇LDOS = ∂LDOS_∂ε .+  ∂LDOS_∂ω .* real.(∂ω_∂ε)

            grad .= ∇LDOS
        end

        push!(mod_LDOS_vals, LDOS)
        push!(mod_omegas, ω₀)
        @show iter, LDOS
        iter += 1
        return LDOS
    end

    function freq_constraint(ρ, grad)
        ε = (max_mat + mat_loss * im) * ρ .+ 1
        ω₀, ∂ω_∂ε = Eigengradient(A, ε, ω, x₀)
        if !isempty(grad) 
            grad .= -real.(∂ω_∂ε)
        end

        return ω - real(ω₀)
    end
    
    ub = vec(mask)
    opt = Opt(:LD_CCSAQ, M * N)
    # opt.params["verbosity"] = 1
    opt.lower_bounds = 0
    opt.upper_bounds = ub
    opt.ftol_rel = ftol
    opt.maxeval = max_eval
    opt.max_objective = mod_LDOS_obj
    # opt.initial_step = 1e-3
    inequality_constraint!(opt, freq_constraint)

    (mod_LDOS_opt, mod_ρ_opt, ret) = optimize(opt, ρ)
    mod_ε_opt = (max_mat + mat_loss * im) * mod_ρ_opt .+ 1
    A_opt, _, _ = Maxwell2d(Lx, Ly, mod_ε_opt, ω; resolution)
    ω₀_opt = sqrt(Arnoldi_eig(A_opt, vec(mod_ε_opt), ω, vec(b))[1])
    Q_opt = -real(ω₀_opt) / 2imag(ω₀_opt)

    @show numevals = opt.numevals # the number of function evaluations
    @show ω₀_opt
    @show Q_opt
    
    return mod_LDOS_opt, reshape(mod_ε_opt, N,M), mod_LDOS_vals, mod_omegas, x, y
end
```

```julia
function top_mask(N, m)
    mask = zeros(Float64, N, N)
    mask[1:m, :] .= 1
    return mask
end
```

```julia
function succ_ref2d(Lx, L₀, ρ_init, ω, b, x₀; resolution=res, max_eval=50, step=1, ftol=0)
    total_design_size = round(Int, resolution * 3)
    start_design_size = round(Int, resolution * L₀)
    mask = top_mask(total_design_size, start_design_size)
    
    _, ε, LDOS_vals, omegas, x, y = LDOS_Optimize_with_mask(Lx, Lx, ρ_init, ω, b, pad_matrix(mask, 3, 3, 0; resolution=res); resolution=resolution, max_eval=max_eval, ftol=ftol, max_mat=11, mat_loss=0)
    
    start_design_size = round(Int,start_design_size + step
    mask = top_mask(total_design_size, start_design_size)
    _, mod_ε_opt, mod_LDOS_vals, mod_omegas, _, _ = mod_LDOS_Optimize_with_mask(Lx, Lx, real(ε .- 1) / 11, ω, b, x₀, pad_matrix(mask, 3, 3, 0; resolution=res); resolution=resolution, max_eval=max_eval, ftol=ftol, max_mat=11, mat_loss=0)
    append!(LDOS_vals, mod_LDOS_vals)
    append!(omegas, mod_omegas)
    
    n = 2
    while L₀ + (n-1) * step < Lx
        start_design_size += step
        mask = top_mask(total_design_size, start_design_size)
        _, mod_ε_opt, mod_LDOS_vals, mod_omegas, x, y = mod_LDOS_Optimize_with_mask(Lx, Lx, real(mod_ε_opt .- 1) / 11, ω, b, x₀, pad_matrix(mask, 3, 3, 0; resolution=res); resolution=resolution, max_eval=max_eval, ftol=ftol, max_mat=11, mat_loss=0)
        append!(LDOS_vals, mod_LDOS_vals)
        append!(omegas, mod_omegas)
        n += 1
    end

    return mod_ε_opt, LDOS_vals, omegas
end
```
